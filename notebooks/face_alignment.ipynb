```python
import face_alignment
from skimage import io
import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
fa = face_alignment.FaceAlignment(face_alignment.LandmarksType._3D, flip_input=False, face_detector="sfd")
# fa = face_alignment.FaceAlignment(face_alignment.LandmarksType._2D, flip_input=False, face_detector="sfd")
```

```python
input = io.imread("../assets/photo/self.png")
input = input[:,:,:3]
pred = fa.get_landmarks(input)      # [0]
```

```python
np.save("../data/landmarks/self_3d.py", pred)
```
